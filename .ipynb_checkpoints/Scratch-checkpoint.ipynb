```python
df['AttendingPhysician'] = df['AttendingPhysician'].fillna('0')
df['AttendingPhysician'] = df['AttendingPhysician'].astype(str)
```

```python
df['BeneID_numeric'] = df['BeneID'].apply(lambda x: int(re.sub('[^0-9]', '', x)))
df['ClaimID_numeric'] = df['ClaimID'].apply(lambda x: int(re.sub('[^0-9]', '', x)))
df['Provider_numeric'] = df['Provider'].apply(lambda x: int(re.sub('[^0-9]', '', x)))
df['Physician_numeric'] = df['AttendingPhysician'].apply(lambda x: int(re.sub('[^0-9]', '', x)))
```

```python
df
```

```python
cols_to_convert = ['ClmAdmitDiagnosisCode','DiagnosisGroupCode', 'ClmDiagnosisCode_1', 'ClmDiagnosisCode_2', 'ClmDiagnosisCode_3', 'ClmDiagnosisCode_4', 'ClmDiagnosisCode_5', 'ClmDiagnosisCode_6', 'ClmDiagnosisCode_7', 'ClmDiagnosisCode_8', 'ClmDiagnosisCode_9', 'ClmDiagnosisCode_10', 'ClmProcedureCode_1', 'ClmProcedureCode_2', 'ClmProcedureCode_3', 'ClmProcedureCode_4', 'ClmProcedureCode_5','ClmProcedureCode_6']
```

```python
def enumerate_values(df, cols_to_convert):

    value_to_int = {}

    for col in cols_to_convert:

        unique_values = df[col].unique()

        for value in unique_values:

            if value not in value_to_int:

                value_to_int[value] = len(value_to_int) + 1
 
    for col in cols_to_convert:

        df[col] = df[col].map(value_to_int)
 
    return df
df = enumerate_values(df, cols_to_convert)
```

```python
df
```

```python
numeric_df = df.select_dtypes(include=['number'])  
numeric_df = numeric_df[['ClaimID_numeric'] + [col for col in numeric_df.columns if col != 'ClaimID_numeric']]  
numeric_df = numeric_df.drop('DeductibleAmtPaid', axis=1) 
numeric_df = numeric_df.drop('ClmProcedureCode_6', axis=1)  
correlation_matrix = numeric_df.corr()  

correlation_matrix
```
