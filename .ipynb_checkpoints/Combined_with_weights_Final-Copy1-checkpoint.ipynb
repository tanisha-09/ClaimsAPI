```python
train_indf=pd.read_csv('Health Claims/Train_Inpatientdata.csv')
train_outdf=pd.read_csv('Health Claims/Train_Outpatientdata.csv')

test_indf=pd.read_csv('Health Claims/Test_Inpatientdata.csv')
test_outdf=pd.read_csv('Health Claims/Test_Outpatientdata.csv')

train_ben=pd.read_csv('Health Claims/Train_Beneficiarydata.csv')
test_ben=pd.read_csv('Health Claims/Test_Beneficiarydata.csv')
```

```python
# Define weights for chronic conditions
chronic_condition_weights = {
    'ChronicCond_Alzheimer': 1,
    'ChronicCond_Heartfailure': 1,
    'ChronicCond_KidneyDisease': 1,
    'ChronicCond_Cancer': 1,
    'ChronicCond_ObstrPulmonary': 1,
    'ChronicCond_Depression': 1,
    'ChronicCond_Diabetes': 1,
    'ChronicCond_IschemicHeart': 1,
    'ChronicCond_Osteoporasis': 1,
    'ChronicCond_rheumatoidarthritis': 1,
    'ChronicCond_stroke': 1
}

# Calculate weighted sum for chronic conditions
def calculate_weighted_chronic_conditions(row):
    weighted_sum = 0
    for col, weight in chronic_condition_weights.items():
        weighted_sum += row[col] * weight
    return weighted_sum

# Apply the function to create the weighted column
final_train['ChronicCondScore'] = final_train.apply(calculate_weighted_chronic_conditions, axis=1)
final_test['ChronicCondScore'] = final_test.apply(calculate_weighted_chronic_conditions, axis=1)
```

```python
cols_to_convert = ['ClmDiagnosisCode_1', 'ClmDiagnosisCode_2', 'ClmDiagnosisCode_3', 'ClmDiagnosisCode_4', 'ClmDiagnosisCode_5', 'ClmDiagnosisCode_6', 'ClmDiagnosisCode_7', 'ClmDiagnosisCode_8', 'ClmDiagnosisCode_9', 'ClmDiagnosisCode_10', 'ClmProcedureCode_1', 'ClmProcedureCode_2', 'ClmProcedureCode_3', 'ClmProcedureCode_4', 'ClmProcedureCode_5','ClmProcedureCode_6']
```

```python
def enumerate_values(final_train, cols_to_convert):
    value_to_int = {}
    for col in cols_to_convert:
        unique_values = final_train[col].unique()
        for value in unique_values:
            if pd.notnull(value) and value not in value_to_int:
                value_to_int[value] = len(value_to_int) + 1
    value_to_int[np.nan] = 0  # Assigning 0 to NaN

    for col in cols_to_convert:
        final_train[col] = final_train[col].map(value_to_int)
 
    return final_train

final_train = enumerate_values(final_train, cols_to_convert)
```

```python
def enumerate_values(final_test, cols_to_convert):
    value_to_int = {}
    for col in cols_to_convert:
        unique_values = final_test[col].unique()
        for value in unique_values:
            if pd.notnull(value) and value not in value_to_int:
                value_to_int[value] = len(value_to_int) + 1
    value_to_int[np.nan] = 0  # Assigning 0 to NaN

    for col in cols_to_convert:
        final_test[col] = final_test[col].map(value_to_int)
 
    return final_test

final_test = enumerate_values(final_test, cols_to_convert)
```

```python
diagnosis_weights = {
    'ClmDiagnosisCode_1': 1,
    'ClmDiagnosisCode_2': 1,
    'ClmDiagnosisCode_3': 1,
    'ClmDiagnosisCode_4': 1,
    'ClmDiagnosisCode_5': 1,
    'ClmDiagnosisCode_6': 1,
    'ClmDiagnosisCode_7': 1,
    'ClmDiagnosisCode_8': 1,
    'ClmDiagnosisCode_9': 1,
    'ClmDiagnosisCode_10': 1
}

procedure_weights = {
    'ClmProcedureCode_1': 1,
    'ClmProcedureCode_2': 1,
    'ClmProcedureCode_3': 1,
    'ClmProcedureCode_4': 1,
    'ClmProcedureCode_5': 1,
    'ClmProcedureCode_6': 1
}

# Calculate score for each row
final_train['DiagnosisScore'] = (final_train.filter(regex='ClmDiagnosisCode_*')
                          .apply(lambda x: x * diagnosis_weights.get(x.name, 0))
                          .sum(axis=1, skipna=True))

final_train['ProcedureScore'] = (final_train.filter(regex='ClmProcedureCode_*')
                          .apply(lambda x: x * procedure_weights.get(x.name, 0))
                          .sum(axis=1, skipna=True))
final_test['DiagnosisScore'] = (final_test.filter(regex='ClmDiagnosisCode_*')
                          .apply(lambda x: x * diagnosis_weights.get(x.name, 0))
                          .sum(axis=1, skipna=True))

final_test['ProcedureScore'] = (final_test.filter(regex='ClmProcedureCode_*')
                          .apply(lambda x: x * procedure_weights.get(x.name, 0))
                          .sum(axis=1, skipna=True))
```
